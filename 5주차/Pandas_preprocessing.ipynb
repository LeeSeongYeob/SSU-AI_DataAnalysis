# 데이터 전처리
- NaN 값 확인
    - info()
    - isnull(), isna()
- NaN 삭제
    - dropna(), dropna(axis=1)
    - dropna(subset=['열이름'])
- NaN 다른 데이터로 대체
    - fillna()

```
import pandas as pd
```

```
train_data = pd.read_csv("./train.csv")
train_data.head()
```

```
# 누락 데이터 확인
train_data.info()
```

```
# 누락된 데이터 True 표시
train_data.isnull()
```

```
# NaN이 있는 데이터 행단위로 삭제 axis 로 행과 열 선택
train_data.dropna()
```

```
# 특정 열 지정해서 삭제
train_data.dropna(subset=['Age'])
```

```
# 특정 열의 NaN의 값을 주어진 값으로 채움
train_data['Age'].fillna(train_data['Age'].mean())
```
