# Matplotlib
- 선 그래프 그리기

```
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False 
```

```
data1 = [10,14,19,20,25]
plt.plot(data1)
plt.show()
```

```
import numpy as np
x  = np.arange(-4.5,5,0.5)
y = 2*x ** 2
print(x)
print(y)
plt.plot(x,y)
plt.show()
```

```
x = np.arange(-5,5,0.1)
y1 = 2*x**2
y2 = 5*x + 30
y3 = 20*np.cos(x)**2

plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

#### plt.figure()
-그래프 여러 창으로 그리기

```
plt.plot(x,y1)
plt.figure()
plt.plot(x,y2)

plt.figure()
plt.plot(x,y3)

plt.show()
```

```
plt.figure(2)
plt.plot(x,y2)

plt.figure(1)
plt.plot(x,y1)
plt.show()
```

#### 그래프 축의 범위 지정하기
- plt.xlim(xmin,xmax)
- plt.ylim(ymin,ymax)

```
x = np.linspace(-4,4,100)
ya = x**3
yb = 10*x**2-2

plt.plot(x,ya,x,yb)
plt.xlim(-1,1)
plt.ylim(-3,3)
plt.show()
```

### 그래프 꾸미기

```
plt.plot(x,ya,'y-',x,yb,'m-.')
plt.show()
```

- 라벨, 제목, 격자, 범례 표시

```
plt.plot(x,ya,'y-',x,yb,'m-')
plt.xlabel('X-axis')    # x축 이름
plt.ylabel('Y-axis')    # y축 이름
plt.title("Graph")      # 이름
plt.grid(True)          # 격자
plt.legend(["data1","data2"])
plt.show()
```

```
x = np.arange(0,6,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
plt.plot(x,y1,'r--^')
plt.plot(x,y2,'s-g')
plt.plot(x,y3,'b:d')
plt.plot(x,y4,'c-.x')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph title")
plt.legend(["data1","data2","data3","data4"])
plt.grid(True)
plt.show()
```
