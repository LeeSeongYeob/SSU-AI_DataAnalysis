```
import pandas as pd
```

```
basket_ball = pd.read_csv("./basketball_stat.csv")
basket_ball.head()
```

### col 이름을 사용하여 맨 마지막에 추가하기

```
basket_ball["3P_double"] = basket_ball["3P"] * 2
basket_ball.head()
```

```
basket_ball['3P_tripple'] = basket_ball['3P_double'] + basket_ball['3P']
basket_ball.head()
```

```
basket_ball.insert(2,'3P10',basket_ball['3P'] * 10)
basket_ball.head()
```

```
basket_ball.loc['new_row'] = 0
basket_ball.tail()
```

```
basket_ball.loc['new_row'] = basket_ball.loc[199] * 2
basket_ball.tail()
```

```
# 열 삭제하기
# inplace 옵션 -> 원본도 삭제
basket_ball.drop(['3P_double','3P_tripple'], axis = 1)
```
