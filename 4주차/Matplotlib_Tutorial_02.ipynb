```
import matplotlib.pyplot as plt
```

```
height = [165,177,160,180,185,155,172]
weight = [62,67,55,74,90,43,64]
plt.scatter(height,weight)
plt.xlabel("Height(m)")
plt.ylabel("Weight(KG)")
plt.title("Height & Weight")
plt.grid(True)
plt.show()
```

```
plt.scatter(height,weight,s=100,c='r',marker='^')
plt.xlabel("Height(m)")
plt.ylabel("Weight(KG)")
plt.title("Height & Weight")
plt.grid(True)
plt.show()
```

```
import numpy as np
size =  100 * np.arange(1,8)
plt.scatter(height,weight,s=size,c=['k','g','b','m','y','c','r'],alpha=0.6)
plt.xlabel("Height(m)")
plt.ylabel("Weight(KG)")
plt.title("Height & Weight")
plt.grid(True)
plt.show()
```

```
cctv = [540,320,600,530,370,540,360,410,380,320,240,480,160,130,400]
people = [3200,800,900,2000,800,1500,1800,1300,900,1200,2000,2100,1600,1000,900]
plt.scatter(cctv,people)
plt.grid(True)
plt.xlabel("cctv")
plt.ylabel("people")
plt.title('relation cctv between people')
plt.show()
```

## 막대 그래프 그리기
-plt.bar(x, height [,width=width_f, color=colors, tick_label=tick_labels, align='center'(기본) 혹은 'edge'])


x는 막대그래프가 x축에 표시될 위치, heigh는 막대 그래프 데이터

```
number_ID = ['m_01','m_02','m_03','m_04']
before_ex = [30,40,42,35]
after_ex = [27,35,40,33]
```

```
n_data = len(number_ID)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.show()
```

```
colors = ['r','g','b','m']
plt.bar(index, before_ex, width = 0.6, color = colors, tick_label = number_ID)
# tick id 는 각각 막대 그래프 이름 지정
plt.show()
```

```
plt.barh(index,before_ex,color = colors,tick_label = number_ID)
plt.show()
```

```
barWidth = 0.4
plt.bar(index, before_ex, color='c', align='edge', width = barWidth)
plt.bar(index+barWidth,after_ex,color='m',align='edge',width=barWidth)

plt.legend(['before','after'])
plt.xlabel('ID')
plt.ylabel('count')
plt.title('diff')
plt.show()
```
